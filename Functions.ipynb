```python
def propogate_orbit_4_plot(M, seconds_diff, a, e, i, cap_omega, omega):

    P = math.pi*2*math.sqrt(a**3/mu)
    print('Period =', P/(60*60), 'hours')
    prop_time_sec = P #priod of orbit
    omega_e = 7.2921158553e-5 #rad/sec
    e_earth = .081819221456 

    M_0 = M
    delta_t = 0

    M_0 = M_0 + n*seconds_diff

    longitude_list = []
    latitude_gd_list = []
    range_list = np.array([])

    while delta_t <= prop_time_sec:
        #step a
        M = M_0 + n*delta_t

        #step b
        E, nu = calc_Eccentric_and_True_Anomaly(M, e)

        #step c
        r_eci, v_eci = Kep_2_Cart.COE2RV(a, e, i, cap_omega, omega, nu, mu)
        range_list = np.append(range_list, np.linalg.norm(r_eci))

        #step d
        hour = 1 + delta_t/(60*60) #hours (float) since midnight UT
        jd = calc_julian_date(2014, 10, 11, hour)#date of interest in UT

        T_UT = (jd - 2451545)/36525 #calc T_UT at this delta_t/date
        #calculate theta gmst @ 0 hr using the T_UT and the eq. from Vallado
        theta_gmst_0_hr = math.radians(100.4606184 + 36000.77005361*T_UT + \
                                       .00038793*T_UT**2 - 2.6e-8*T_UT**3)

        seconds = hour * 60 * 60 #seconds since midnight UT
        #calculate theta gmst using theta gmst @ 0 hr and seconds 
        #since midnight*the rotation rate of earth
        theta_gmst = theta_gmst_0_hr + omega_e * seconds
        theta_gmst = theta_gmst % (2*math.pi)

        #step e
        r_ecef = eci2ecef(r_eci, theta_gmst)

        #step f
        latitude, longitude, altitude = ecef2geo_lat_lon_alt(r_ecef, earth_radius)

        #step g
        latitude_geodetic = math.atan2(math.tan(latitude), 1-e_earth**2)

        longitude_list.append(longitude)
        latitude_gd_list.append(latitude_geodetic)

        delta_t += 60

    lon_array = np.array(longitude_list)
    lat_gd_array = np.array(latitude_gd_list)
    
    range_max = np.amax(range_list)
    
    return (lon_array, lat_gd_array, range_max)


def determine_orbital_regime(range_max):
    earth_radius = 6378.1363
    alt = range_max - earth_radius
    
    if alt >= 80 and alt <= 2000:
        regime = 'LEO'
    elif alt > 2000 and alt <= 35786:
        regime = 'MEO'
    elif alt > 35786:
        regime = 'GEO'
    return regime
    
#input the lat, lon, alt, (radians) and radius of main body
#in order to calculate the ecef x, y, z position of a ground sensor/position
def topo2ecef(latitude, longitude, altitude, main_body_radius):
    r = altitude + main_body_radius
    x_ecef = r*math.cos(latitude)*math.cos(longitude)
    y_ecef = r*math.cos(latitude)*math.sin(longitude)
    z_ecef = r*math.sin(latitude)
    r_ecef = np.array([x_ecef, y_ecef, z_ecef])
    return r_ecef

#propogation equations
def orbitpropogator(t, R_V): 
    R, V = R_V.reshape(2, 3)
    r = np.linalg.norm(R)
    a_vec = (-mu/r**3) * R
    dx = [V, a_vec]
    dx = np.array(dx)
    dx = dx.flatten()
    return dx

def propagate_4_xyz_plot(r, v, time_plot):
    #initializations for integrator
    X_V = np.append(r, v)
    y0 = X_V
    y0 = y0.flatten()
    t_init =  0
    t_final = time_plot
    t_step = 10 

    integrator = ode(orbitpropogator)
    integrator.set_integrator('dopri5', rtol=1e-4, atol=1e-4)
    integrator.set_initial_value(y0, t_init)

    result = []
    result.append(np.insert(y0, 0, t_init))

    while integrator.successful() and integrator.t < t_final:
        integrator.integrate(integrator.t + t_step)

        result.append(np.insert(integrator.y, 0, integrator.t))

    result = np.array(result)

    integrator_times = result[:, 0]
    X = result[:, [1, 2, 3]]

    x = X[:, 0]
    y = X[:, 1]
    z = X[:, 2]
    return (x, y, z)
```
